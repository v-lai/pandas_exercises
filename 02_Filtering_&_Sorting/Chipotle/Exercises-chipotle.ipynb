### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')
```

```python
to_float = lambda x: float(x[1:-1])
chipo.item_price = chipo.item_price.apply(to_float)
chipo.head()
```

### Step 4. How many products cost more than $10.00?

```python
chipo_filtered = chipo.drop_duplicates(['item_name', 'quantity', 'choice_description'])
# chipo_filtered # 1949
chipo_quantity_1 = chipo_filtered[chipo_filtered.quantity == 1]
# chipo_quantity_1 # 1806
chipo_price_10_plus = chipo_quantity_1[chipo_quantity_1['item_price'] > 10]
chipo_price_10_plus # 685
chipo_price_10_plus.item_name.nunique() # 25

greater_than_10 = chipo.query('item_price > 10').item_name.nunique() # 31
greater_than_10
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo_filter_name_quantity = chipo.drop_duplicates(['item_name', 'quantity'])
chipo_filter_quantity_1 = chipo_filter_name_quantity[chipo_filter_name_quantity.quantity == 1]
chipo_filter_quantity_1[['item_name', 'item_price']]
```

### Step 6. Sort by the name of the item

```python
chipo_filter_quantity_1[['item_name', 'item_price']].sort_values(by='item_name')
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
most_expensive = chipo.sort_values(by='item_price', ascending=False).head(1)
most_expensive
# most_expensive['quantity']
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
veggie_salad_bowls = chipo.query('item_name == "Veggie Salad Bowl"')
# veggie_salad_bowls
veggie_salad_bowls['quantity'].sum()
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
one_canned_soda = chipo.query('item_name == "Canned Soda" & quantity == 1') 
# one_canned_soda
one_canned_soda['quantity'].count()
```
